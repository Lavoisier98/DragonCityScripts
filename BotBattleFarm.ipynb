```python
import pyautogui as pg
import pyperclip as pc
import time as tm

def clickOnTime(*parametros):
#Ordem: local, ação, confiança, Left, Top, Width, Height, numeroDeWidths, numeroDeHeights, timeBefore, timeAfter, grayscale
    tela = pg.size()
    wayStart = r"C:\Users\user\Desktop\Felipe\Jupyter\PNGs"
    wayEnd = r".png"
    quantDeParametros = len(parametros)
    #Grayscale para agilizar
    if quantDeParametros < 12 or parametros[11] == None:
        funcGrayScale = True
    else:
        funcGrayScale = parametros[11]

    #Número de widths e heights de distância do centro da figura
    if quantDeParametros < 8 or parametros[7] == None:
        numeroDeWidths = 0.5
    else:
        numeroDeWidths = parametros[7]
    if quantDeParametros < 9 or parametros[8] == None:
        numeroDeHeights = 0.5
    else:
        numeroDeHeights = parametros[8]
        
    #Timeout antes e depois do clique
    if quantDeParametros < 11 or parametros[10] == None:
        timeOutBefore = 0.1
    if quantDeParametros < 10 or parametros[9] == None:
        timeOutAfter = 0.1

    #Region
    if quantDeParametros < 4 or parametros[3] == None:
        posLeft = 0
    else:
        if int(parametros[3])>1:
            posLeft = int(parametros[3])
        else:
            posLeft = int(parametros[3]*tela.width)
    if quantDeParametros < 5 or parametros[4] == None:
        posTop = 0
    else:
        if int(parametros[4])>1:
            posTop = int(parametros[4])
        else:
            posTop = int(parametros[4]*tela.height)
    if quantDeParametros < 6 or parametros[5] == None:
        posWidth = tela.width-1-posLeft
    else:
        if int(parametros[5])>1:
            posWidth = int(parametros[5])-posLeft
        else:
            posWidth = int(parametros[5]*tela.width)-posLeft
    if quantDeParametros < 7 or parametros[6] == None:
        posHeight = tela.height-1-posTop
    else:
        if int(parametros[6])>1:
            posHeight = int(parametros[6])-posTop
        else:
            posHeight = int(parametros[6]*tela.height)-posTop
        
    #Basics: local, ação, confiança
    if quantDeParametros < 2 or parametros[1] == None:
        conf = 0.8
    else:
        conf = parametros[1]
    if quantDeParametros < 3 or parametros[2] == None:
        action = "Clicar"
    else:
        action = parametros[2]
    wayConcat = wayStart+parametros[0]+wayEnd
    localDaImagem = None
    while localDaImagem == None:
        localDaImagem = pg.locateOnScreen(wayConcat, confidence=conf, grayscale=funcGrayScale, region=(posLeft, posTop, posWidth, posHeight))
        if localDaImagem == None:
            conf -= 0.005
            print(conf)
    xReturn = localDaImagem.left + numeroDeWidths * localDaImagem.width
    yReturn = localDaImagem.top + numeroDeHeights * localDaImagem.height
    tm.sleep(timeOutBefore)
    pg.moveTo(pg.position(x=xReturn,y=yReturn))
    if action == "Clicar":
        tm.sleep(0.1)
        pg.click(pg.position(x=xReturn,y=yReturn))
        tm.sleep(timeOutAfter)
tm.sleep(3)
way1 = r"\Quest1"
#Ordem: local, confiança, ação, %Left, %Top, %Width, %Height, numeroDeWidths, numeroDeHeights, timeBefore, timeAfter, grayscale
clickOnTime(way1,0.96,"Achar")
```
