```python
import time as tm
import pyautogui as pg

wayStart = r"C:\Users\user\Desktop\Felipe\Jupyter\PNGs"
wayEnd = r".png"
tela = pg.size()

def localizar(selectedWay, functionGrayScale, Confidence, Action, Region):
    Imagem = wayStart + selectedWay + wayEnd
    localDaImagem = pg.locateOnScreen(Imagem, grayscale=functionGrayScale, confidence=Confidence, region=(Region[0],Region[1], Region[2], Region[3]))
    if localDaImagem != None:
        if Action == "Clicar":
            clickOnTime(localDaImagem, 0.2, 0)
        return localDaImagem

def clickOnTime(local, timeOutBefore, timeOutAfter):
    tm.sleep(timeOutBefore)
    pg.moveTo(local)
    tm.sleep(0.1)
    pg.click(local)
    tm.sleep(timeOutAfter)
    
way1 = [r"\HeroicRunFight1",0.86,0,int(0.651*tela.height),tela.width-1,int(0.208*tela.height)-1]
way2 = [r"\HeroicRunFight2",0.94,0,0,tela.width-1,tela.height-1]
way3 = [r"\HeroicRunFight3",0.73,0,0,tela.width-1,tela.height-1]
way4 = [r"\HeroicRunFight4",0.89,0,0,tela.width-1,tela.height-1]
way5 = [r"\HeroicRunFight5",0.81,0,0,tela.width-1,tela.height-1]

tm.sleep(4)
wayA = way1
wayB = way5
fracassos = 0

while True:
    #1° Passo: Esperar aparecer em tela o Fight Button e clicá-lo
    tm.sleep(0.5)
    x1 = None
    x2 = None
    fracassos = 0
    while x1 == None:
        fracassos = 0
        while x1 == None and fracassos < 3:
            x1 = localizar(wayA[0], True, wayA[1],"Clicar",[wayA[2],wayA[3],wayA[4],wayA[5]])
            if x1 == None:
                x2 = localizar(wayB[0], True, wayB[1],"Clicar",[wayB[2],wayB[3],wayB[4],wayB[5]])
                if x2 == None:
                    if wayA[0] == way1[0]:
                        tm.sleep(7)
                    fracassos += 1
        if fracassos >= 3 and (wayA[0] == way1[0] or wayB[0] == way1[0]):
            #Prevenir proteção de tela
            pg.moveTo(int(0.9*tela.width),int(0.1*tela.height),3)
            pg.click(int(0.9*tela.width),int(0.1*tela.height))
            pg.moveTo(int(0.9*tela.width),int(0.9*tela.height),3)
            pg.moveTo(int(0.1*tela.width),int(0.9*tela.height),3)
            pg.moveTo(int(0.1*tela.width),int(0.1*tela.height),3)
            fracassos = 0
    fracassos = 0
    if wayA[0] == way1[0]:
        wayA = way2
        wayB = way1
    elif wayA[0] == way2[0]:
        wayA = way3
        wayB = way2
    elif wayA[0] == way3[0]:
        wayA = way4
        wayB = way3
    elif wayA[0] == way4[0]:
        wayA = way5
        wayB = way4
    elif wayA[0] == way5[0]:
        wayA = way1
        wayB = way5
        

    
```
